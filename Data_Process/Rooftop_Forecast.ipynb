```python
import requests
from bs4 import BeautifulSoup
import zipfile
import os
import time
```

```python


# Fetch the webpage
url = "https://nemweb.com.au/Reports/Archive/ROOFTOP_PV/FORECAST/"
response = requests.get(url)

# Parse the HTML
soup = BeautifulSoup(response.text, 'html.parser')

# Find all links to ZIP files
zip_links = [a['href'] for a in soup.find_all('a') if a['href'].endswith('.zip')]

for link in zip_links:
    # Correct the URL construction
    full_url = "https://nemweb.com.au" + link
    file_name = link.split('/')[-1]
    
    base = "./Roof_Forecast"
    file_path = os.path.join(base, file_name)
    print(f"Attempting to download: {full_url}")
    
    r = requests.get(full_url, stream=True)

    with open(file_path, 'wb') as f:
        for chunk in r.iter_content(chunk_size=8192):
            f.write(chunk)
    
    
        
        
    
```

```python
from zipfile import ZipFile
files = os.listdir("./Roof_Forecast")
zip_files = [f for f in files if f.endswith('.zip')]
zip_files.sort()
print(zip_files)


for file in zip_files:
    file = os.path.join("./Roof_Forecast" , file)
    date_str = file.split('_')[-1].replace('.zip', '')

    new_dir = f"./Roof_sub/{date_str}"
    os.makedirs(new_dir,exist_ok=True)
    with zipfile.ZipFile(file, 'r') as zip_ref:
        zip_ref.extractall(new_dir)
    
```

```python
import pandas as pd
import numpy as np

base = "./Roof_sub"
csvfolders = os.listdir(base)
csvfolders = [f for f in csvfolders if "csv" in f]
csvfolders.sort()
states = ["NSW1","QLD1","VIC1","TAS1","SA1"]
time_cols = ["0.5 hour future","1 hour future", "1.5 hours future", "2 hours future", 
                "2.5 hours future", "3 hours future", "3.5 hours future", "4 hours future",
                "4.5 hours future", "5 hours future", "5.5 hours future", "6 hours future",
                "6.5 hours future", "7 hours future", "7.5 hours future", "8 hours future",
                "8.5 hours future", "9 hours future", "9.5 hours future", "10 hours future"]

for csvfolder in csvfolders:
    folderpath = os.path.join(base,csvfolder)
    csvfiles = os.listdir(folderpath)
    csvfiles.sort()
    print(csvfolder)
    for file in csvfiles:
        folder_path = os.path.join(base,csvfolder)
        file_path = os.path.join(folder_path,file)
        
        

        df = pd.read_csv(file_path,skiprows=1, header=0,index_col=False)
        
        time_cur = df["VERSION_DATETIME"][11]
        
        for state in states:
            new_df = pd.DataFrame(columns=["date","Region","0.5 hour future","1 hour future", "1.5 hours future", "2 hours future", 
                "2.5 hours future", "3 hours future", "3.5 hours future", "4 hours future",
                "4.5 hours future", "5 hours future", "5.5 hours future", "6 hours future",
                "6.5 hours future", "7 hours future", "7.5 hours future", "8 hours future",
                "8.5 hours future", "9 hours future", "9.5 hours future"])
            
            next_index = len(new_df)
            new_df.loc[next_index, 'date'] = time_cur
            new_df.loc[next_index, 'Region'] = state

            new_df["date"] = pd.to_datetime(new_df["date"])
            new_df["date"] = new_df["date"].dt.round("30min")

            temp_df = df[df["REGIONID"]==f"{state}"].iloc[:20]
            temp_df = temp_df[["REGIONID","INTERVAL_DATETIME","POWERMEAN","POWERPOE50"]]
            
            for i, label in enumerate(time_cols):
                new_df[[f'{label}']] = temp_df.iloc[i]['POWERMEAN']
                new_df[[f'{label}']] = temp_df.iloc[i]['POWERPOE50']
            
            if os.path.isfile("Roof_Forecast.csv"):

                new_df.to_csv("Roof_Forecast.csv", mode='a', header=False, index=False)
            else:
                new_df.to_csv("Roof_Forecast.csv", mode='w', header=True, index=False)
        
    
    
```
